```
                "toDate":"202005070000"
                }'

```






````
# Necessary imports
import json
import pandas as pd
import os
````

````
final_df = pd.DataFrame()

# Path contains text files in JSON format, containing the result of API call
path = "TweetFiles/Tesla/"

for filename in os.listdir(path):
    with open(path + filename) as file:
        js = json.loads(file.read())['results']
        df = pd.DataFrame(data=js)
        df = df[['created_at', 'id', 'text']]
        final_df = final_df.append(df, ignore_index=True)

# drop duplicate tweets and then sort the final dataframe
final_df = final_df.drop_duplicates(['text']).sort_values(by='id').reset_index(drop=True)

final_df.to_csv(PATH + "tesla_tweets.csv")
````
